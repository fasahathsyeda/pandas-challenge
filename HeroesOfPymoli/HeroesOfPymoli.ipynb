## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players

```python
duplicate_drop= purchase_data.drop_duplicates(["SN"])
duplicate_drop.shape
gender_group = duplicate_drop.groupby(["Gender"])
gender_group.head(10)
```

```python
gender_group_df= pd.DataFrame()
gender_group_df["Gender"]=gender_group["Gender"].count()
gender_group_df
gender_group_df = gender_group_df.rename(columns={"Gender": "Total Count"}).reset_index()

gender_group_df["Percentage of Players"]=(gender_group_df["Total Count"]/len(total_players_df)).map("{:.2%}".format)

gender_group_df.style.hide_index()
```
