```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.columns
```

## Player Count

```python
total_unique_players=purchase_data["SN"].unique()
total_unique_players
Unique_players_df=pd.DataFrame({"Total Players": [len(total_unique_players)]})
Unique_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
unique_items=purchase_data["Item Name"].unique()
len(unique_items)
average_price=purchase_data["Price"].mean()
average_price
number_purchases=purchase_data["Item ID"].count()
number_purchases
total_revenue=purchase_data["Price"].sum()
total_revenue
pd.DataFrame({"Number of Unique Items" : [len(unique_items)],"Average Price" : [ str('${:.2f}'.format(average_price))],"Number of Purchases" : [number_purchases],"Total Revenue" : [ str('${:.2f}'.format(total_revenue))]})
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
duplicate_drop= purchase_data.drop_duplicates(["SN"])
duplicate_drop.shape
gender_group_unique = duplicate_drop.groupby(["Gender"])
gender_group_unique.head(10)
```

```python
gender_demographics= pd.DataFrame()
gender_demographics["Gender"]=gender_group_unique["Gender"].count()
gender_demographics
gender_demographics = gender_demographics.rename(columns={"Gender": "Total Count"}).reset_index()
gender_demographics
gender_demographics["Percentage of Players"]=(gender_demographics["Total Count"]/len(total_unique_players)).map("{:.2%}".format)
gender_demographics.style.hide_index()
```
